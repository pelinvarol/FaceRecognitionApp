### Feature Extraction with Eigen Face

```python
# new face - mean face - eigen face
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

# eigen face
from sklearn.decomposition import PCA
import pickle
```

```python
# Load the data
data = pickle.load(open('./data/data_images_100_100.pickle', mode = 'rb'))
```

```python
data.head()
```

#### Eigen Faces

```python
# Mean face
# split data into two parts: independent variable and dependent variable, drop the gender
X = data.drop('gender', axis = 1).values # all images
```

```python
X
```

```python
X.shape
```

```python
mean_face = X.mean(axis = 0) #we want all the pixels, set the axis equal to zero.
```

```python
mean_face.shape #10000 pixels
```

```python
# visualize mean face
plt.imshow(mean_face.reshape((100,100)), cmap='gray')
plt.axis('off')
plt.show()
```

 **Subtract data with mean face**

```python
X_t = X - mean_face # transformed data
```

```python
# what is a number of components that we need to use?
# to know the answer we need to use the elbow method
```

**Apply X_t data to PCA**
- Find the right number of components - Elbow

- With right number of components, compute Principal components

```python
pca = PCA(n_components=None, whiten=True, svd_solver='auto')
pca.fit(X_t)
```

```python
# explained variance - cumulative explained variance
exp_var_df = pd.DataFrame()
exp_var_df['explained_var'] = pca.explained_variance_ratio_
exp_var_df['cum_explained_var'] = exp_var_df['explained_var'].cumsum()
exp_var_df['principal_components'] = np.arange(1, len(exp_var_df)+1)

exp_var_df.head()
```

```python
exp_var_df.set_index('principal_components', inplace=True)
```

```python
# visualize explained variance
fig, ax = plt.subplots(nrows=2, figsize=(15,12))

exp_var_df['explained_var'].head(100).plot(kind='line', marker='o', ax = ax[0])
exp_var_df['cum_explained_var'].head(100).plot(kind='line', marker='o', ax = ax[1])
```

** From above plot, it's clear that with approximately 50 principal components having 80% converage. **

```python
pca_50 = PCA(n_components=50, whiten=True, svd_solver='auto')
pca_data = pca_50.fit_transform(X_t)
```

```python
# we have successfully train our PCA with 50 components and also we got the pca_data
pca_data.shape
# this is the data that we're going to use in the machine learning model process
```

```python
# SAVING DATA AND MODELS
y = data['gender'].values # independent variables
np.savez('./data/data_pca_50_target', pca_data, y)
```

```python
# saving the model
pca_dict = {'pca': pca_50, 'mean face':mean_face}
```

```python
pickle.dump(pca_dict, open('model/pca_dict.pickle', 'wb'))
```

#### Visualize Eigen Image

```python
# in order to visualize eigen image, we're going to use pca_data 
 #  pca_data.shape -> (4320, 50)
# we need to inverse transform this
pca_data_inv = pca_50.inverse_transform(pca_data)
pca_data_inv.shape
```

```python
eigen_image = pca_data_inv[0,:].reshape((100,100))
```

```python
eigen_image.shape
```

```python
plt.imshow(eigen_image,cmap='gray')
plt.axis('off')
plt.show()
```
