```python
import cv2
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
# female path
fpath = glob('./data/female/*.jpg') # this will return all images path in a list
# male path
mpath = glob('./data/male/*.jpg')
```

```python
# !!! this is how we can extract all the path of the images in any particular folder !!!
fpath
```

```python
print('The number of images in female folder is', len(fpath))
```

```python
print('The number of images in male folder is', len(mpath))
```

```python
### STEP 1: Read Image and Convert to RGB
img = cv2.imread(fpath[0]) #read image in bgr format - cause opencv does that
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

### STEP 2: Apply Haar Cascade Classifier
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
faces_list = haar.detectMultiScale(gray, 1.5, 5)
print(faces_list) # we get bounding box information as [[80 86 273 273]]
for x,y,w,h in faces_list:
    cv2.rectangle(img_rgb, (x,y), (x+w, y+h), (0,255,0),2)
    ### STEP 3: Crop Faces
    # roi = region of interest
    # row information -> y:y+h
    #column information -> x: x+w
    roi = img_rgb[y:y+h, x:x+w]
    plt.imshow(roi)
    plt.axis('off')
    plt.show()

### STEP 4: Save Image
```

# Crop all *female* images

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
for i in range(len(fpath)):
    try:
       # if i == 5: # testing for only five iterations just to check - we will have five cropped female data
          #  break
        ### STEP 1: Read Image and Convert to RGB
        img = cv2.imread(fpath[i]) #read image in bgr format - cause opencv does that
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        ### STEP 2: Apply Haar Cascade Classifier
        gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
        faces_list = haar.detectMultiScale(gray, 1.5, 5)

        for x,y,w,h in faces_list:
            ### STEP 3: Crop Faces
            # roi = region of interest
            # row information -> y:y+h
            #column information -> x: x+w
            roi = img[y:y+h, x:x+w]
            ### STEP 4: Save Image
            cv2.imwrite(f'./crop_data/female/female_{i}.jpg', roi)
            print('Image successfully processed!')
    except:
        print('Unable to process the image.')
```
