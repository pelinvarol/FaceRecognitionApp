```python
import numpy as np
import pandas as pd

# visualization
import matplotlib.pyplot as plt
import seaborn as sns

# computer vision library
import cv2

# glob
from glob import glob
```

```python
# Step Number 1: Extract path of female and male in crop_data folder and put them in a list
fpath = glob('./crop_data/female/*.jpg')
mpath = glob('./crop_data/male/*.jpg')
```

```python
# there will be separate dataframes for male and female
df_female = pd.DataFrame(fpath, columns = ['filepath'])
df_female['gender'] = 'female'

df_male = pd.DataFrame(mpath, columns = ['filepath'])
df_male['gender'] = 'male'

# concat two dataframes
# just provide axis is equal to zero so that we can be able to perform the
# union operation just like the SQL joins.
df = pd.concat((df_female, df_male),axis =0)
df.head()
```

```python
# for each and every image, going to extract the width and height of the image
# they're square matrix so we can either take width or height

# it will take each image path, then return width of the image
def get_size(path):
    img = cv2.imread(path)
    return img.shape[0]
```

```python
df['dimension'] = df['filepath'].apply(get_size) #store dimension of image in this columns
```

### Structure the images
- 100x100

```python
def structuring(path):
    try:
        # step 1: read the image
        img = cv2.imread(path) #BGR

        # step 2: convert into grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # step 3: resize into 100 x 100 array
        size = gray.shape[0]
        if size >= 100:
            # cv2.INTER_AREA -> shrinking
            gray_resize = cv2.resize(gray, (100,100), cv2.INTER_AREA)
        else:
            # cv2.INTER_CUBIC -> enlarging
            gray_resize = cv2.resize(gray, (100,100), cv2.INTER_CUBIC)

        # step 4: Flatten Image(1x10,000)
        # its always good to store information in one dimension format.That's the reason flattening process.
        flatten_image = gray_resize.flatten()
        return flatten_image
    except:
        return None
```

```python
pd.set_option('mode.chained_assignment', None)
df_filter['data'] = df_filter['filepath'].apply(structuring) #convert all images into 100x100
```
